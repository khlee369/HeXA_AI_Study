```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

m = 5000
mu = np.array([0, 0])
sigma = np.array([[3, 1.5], 
                  [1.5, 1]])

X = np.random.multivariate_normal(mu, sigma, m)
X = np.asmatrix(X)

fig = plt.figure(figsize=(10, 6))
plt.plot(X[:,0], X[:,1], 'k.')
plt.axis('equal')
plt.show()
```

```python
X.shape
```

```python
m
```

```python
cov = (X.T*X)/(m-1)
```

```python
cov
```

```python
H, E = np.linalg.eig(cov)
```

```python
print(H)
```

```python
print(E)
```

```python
xp = np.linspace(-6,6, 200)
yp = E[1,0]/E[0,0]*xp

fig = plt.figure(figsize=(10, 6))
plt.plot(X[:,0], X[:,1], 'k.')
plt.plot(xp, yp)
plt.axis('equal')
plt.show()
```

```python
plt.plot(X[:,0], tf_X[:,0], 'k.')
plt.show()
```

```python
plt.stem(H/np.sum(H))
plt.xlim([-0.5,1.5])
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
pca.transform(X)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```
