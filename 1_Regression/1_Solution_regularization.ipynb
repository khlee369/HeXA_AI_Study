# Problem

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4.5, 4.5, 10)
y = np.array([0.9819, 0.7973, 1.9737, 0.1838, 1.3180, -0.8361, -0.6591, -2.4701, -2.8122, -6.2512])

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xlim(-5, 5)
plt.ylim(-7, 3)
plt.show()
```

# Solution

```python
# y_hat = theta[2]*x^2 + theta[1]*x + theta[0]
# gradient descent algorithm
theta = np.array([0.0, 0.0, 0.0])
lamb = 1
learning_rate = 7e-7

y_hat = theta[2]*(np.square(x)) + theta[1]*x + theta[0]
loss = np.mean(np.square(y_hat - y)) + lamb*np.sum(np.square(theta))

n_iter = 400000
for i in range(n_iter):
    theta[2] = theta[2] - learning_rate*(np.mean(np.square(x)*(y_hat - y)) + lamb*theta[2])
    theta[1] = theta[1] - learning_rate*(np.mean(x*(y_hat - y)) + lamb*theta[1])
    theta[0] = theta[0] - learning_rate*(np.mean(y_hat - y) + lamb*theta[0])
    
    if i%20000 == 0 :
        y_hat = theta[2]*(np.square(x)) + theta[1]*x + theta[0]
        loss = np.mean(np.square(y_hat - y)) + lamb*np.sum(np.square(theta))
        print("loss : {}".format(loss))

plt.figure(figsize=(10, 6))     
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xlim(-5, 5)
plt.ylim(-7, 3)
plt.plot(x, y_hat)
plt.plot(x, y, 'o')
plt.show()
```

# Solution MatrixForm

```python
X = np.array([x**2, x**1, x**0]).T
Y = y.reshape([-1,1])

theta = np.array([0.0, 0.0, 0.0])
theta = theta.reshape([3,1])
Y_hat = np.matmul(X, theta)

lamb = 1
learning_rate = 7e-7

loss = np.mean(np.square(Y_hat - Y)) + lamb*np.sum(np.square(theta))

n_iter = 400000
for i in range(n_iter):
    grad = np.matmul(np.matmul(X.T, X), theta) - np.matmul(X.T, Y)
    regul = theta
    
    theta = theta - learning_rate*(grad + lamb*regul)
    
    if i%20000 == 0 :
        Y_hat = np.matmul(X, theta)
        loss = np.mean(np.square(Y_hat - Y)) + lamb*np.sum(np.square(theta))
        print("loss : {}".format(loss))

plt.figure(figsize=(10, 6))     
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xlim(-5, 5)
plt.ylim(-7, 3)
plt.plot(x, Y_hat)
plt.plot(x, Y, 'o')
plt.show()
```
